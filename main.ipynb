Depending on the context, a single phrase can mean several things, which could result in misunderstandings. This becomes a serious problem, particularly in the medical field where accuracy is crucial. This will particularly effect in synonyms in which one single word can alter the whole sentence or paragraph. 

**Limited to textual data only:**
NLP is primarily restricted to textual data processing and understanding, despite its great efficacy in this domain. It might miss information from other crucial data modalities, such as sensor data, audio recordings, or medical pictures. This drawback emphasizes the necessity of complementary technology when working with heterogeneous patient data. Sometimes the handwritings of certain individuals will be hard for the model to pick up if it is not trained properly or lacks information of that.

**Data Availability:** NLP applications' timeliness depends on the availability of current textual data. The information processed by NLP is guaranteed to reflect the most recent patient contacts and medical events thanks to real-time access to clinical notes and EHRs. The accuracy of the model will continue to be rather high because there are hundreds of these notes available, and the number is growing every day.

**Setup Time:** Training NLP models on particular medical language subtleties is necessary before using them in healthcare applications. Improvements in pre-trained models and transfer learning can speed up this process, even if it still needs a reasonable amount of initial setup time.

**Time to produce results:** Information is processed by NLP rather quickly after the models are configured and trained. Because of its real-time applicability, it can be used for jobs like clinical decision support, where quick patient record insights are essential for taking appropriate actions.

**Input and outputs**
Input:  medical records, such as clinical notes and electronic health records, are examples of unstructured input.
Output: Structured data, insights, and information extracted for clinical decision assistance.
NLP is a powerful tool for managing the complexity of healthcare data because it provides a link between the rich information contained in textual records and the need for rapid, relevant insights. NLP's advantages make it an essential tool in the toolbox of AI techniques meant to revolutionize healthcare and medicine, even in the face of obstacles.

```python
import tensorflow as tensor

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, BatchNormalization
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
import numpy as numpy
import matplotlib.pyplot as plot
from os import listdir
from os.path import isfile, join
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as seaborn
```

```python
train_path = "C:\\Users\\hp\\Bones\\dataset\\train"
test_path = "C:\\Users\\hp\\Bones\\dataset\\val"
# Simply download the dataset from here https://drive.google.com/drive/folders/1WJc4zfmL-l1MktJ9E6nlumlSxTPBYHJk?usp=sharing and change the paths
```

```python
# Declaring image sizes and dimensions
img_width, img_height = 150, 150
batch_size = 32
epochs = 12
```

```python
# Build the CNN model
CNNmodel = Sequential()

CNNmodel.add(Conv2D(32, (3, 3), input_shape=(img_width, img_height, 3), activation='relu'))
CNNmodel.add(MaxPooling2D(pool_size=(2, 2)))
CNNmodel.add(BatchNormalization())

CNNmodel.add(Conv2D(64, (3, 3), activation='relu'))
CNNmodel.add(MaxPooling2D(pool_size=(2, 2)))
CNNmodel.add(BatchNormalization())

CNNmodel.add(Conv2D(128, (3, 3), activation='relu'))
CNNmodel.add(MaxPooling2D(pool_size=(2, 2)))
CNNmodel.add(BatchNormalization())

CNNmodel.add(Conv2D(256, (3, 3), activation='relu'))
CNNmodel.add(MaxPooling2D(pool_size=(2, 2)))
CNNmodel.add(BatchNormalization())

#This would contain 3 Layers

CNNmodel.add(Flatten())

CNNmodel.add(Dense(1, activation='sigmoid'))
```

```python
# Compile the model
CNNmodel.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# Data augmentation to prevent overfitting
train_data = ImageDataGenerator(rescale=1./255, shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
test_data = ImageDataGenerator(rescale=1./255)

train_gen = train_data.flow_from_directory(train_path, target_size=(img_width, img_height),
                                                    batch_size=batch_size, class_mode='binary')

test_gen = test_data.flow_from_directory(test_path, target_size=(img_width, img_height),
                                                  batch_size=batch_size, class_mode='binary')
```

```python
# Train the model
modelHistory = CNNmodel.fit(train_gen, steps_per_epoch=train_gen.samples // batch_size,
                    epochs=epochs, validation_data=test_gen,
                    validation_steps=test_gen.samples // batch_size)
```

```python
# Plot training History
plot.plot(modelHistory.history['accuracy'], label='accuracy')
plot.plot(modelHistory.history['val_accuracy'], label='val_accuracy')
plot.xlabel('Epoch')
plot.ylabel('Accuracy')
plot.legend()
plot.show()

# Plot accuracy over epochs
plot.plot(modelHistory.history['accuracy'], label='Training Accuracy')
plot.plot(modelHistory.history['val_accuracy'], label='Validation Accuracy')
plot.xlabel('Epoch')
plot.ylabel('Accuracy')
plot.legend()
plot.title('Accuracy Over Epochs')
plot.show()

# Plot loss over epochs
plot.plot(modelHistory.history['loss'], label='Training Loss')
plot.plot(modelHistory.history['val_loss'], label='Validation Loss')
plot.xlabel('Epoch')
plot.ylabel('Loss')
plot.legend()
plot.title('Loss Over Epochs')
plot.show()
```

```python
# Test the model with a new image
def predict_single_fracture(CNNmodel, image_path):
    input_img = image.load_img(image_path, target_size=(img_width, img_height))
    input_img = image.img_to_array(input_img)
    input_img = numpy.expand_dims(input_img, axis=0)
    input_img /= 255.0  # Normalize the image

    result = CNNmodel.predict(input_img)
    if result[0][0] > 0.5:
        print("Fracture detected.")
    else:
        print("No fracture detected.")
```

```python
def predict_folder_accuracy(CNNmodel, folder_path, img_width, img_height):
    image_files = [f for f in listdir(folder_path) if isfile(join(folder_path, f))]
    
    true_labels = []
    predicted_labels = []

    for image_file in image_files:
        image_path = join(folder_path, image_file)

        img = image.load_img(image_path, target_size=(img_width, img_height))
        img = image.img_to_array(img)
        img = img.reshape((1,) + img.shape)
        img /= 255.0

        result = CNNmodel.predict(img)
        predicted_label = 1 if result[0][0] > 0.5 else 0

        true_label = 1 if "fracture" in image_file.lower() else 0

        true_labels.append(true_label)
        predicted_labels.append(predicted_label)

    accuracy = accuracy_score(true_labels, predicted_labels)
    print(f"Accuracy: {accuracy:.2%}")
    
     # Generate confusion matrix
    cm = confusion_matrix(true_labels, predicted_labels)
    print("Confusion Matrix:")
    print(cm)

    # Plot the confusion matrix
    plot.figure(figsize=(4, 4))
    seaborn.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=['No Fracture', 'Fracture'], yticklabels=['No Fracture', 'Fracture'])
    plot.title('Confusion Matrix')
    plot.xlabel('Predicted Label')
    plot.ylabel('True Label')
    plot.show()


# Example usage:
folder_path_to_predict = "C:\\Users\\hp\\Bones\\dataset\\test"
predict_folder_accuracy(CNNmodel, folder_path_to_predict, img_width, img_height)
```

## Testing methods
**Using a Validation set .**
Divide the dataset into sets for validation and training. To evaluate the model's generalization performance, use the validation set after training the model on the training set.This makes it easier to determine whether the model is overfitting the training set.

**Test Set:**

Utilizing a fully independent test set, assess the model's ultimate performance once it has been trained and validated.
Images that the model has never seen before should be included in the test set.

**Performance Metrics:**

To assess the model's performance, use relevant measures like accuracy, precision, recall, F1-score, and confusion matrix.
These metrics offer an all-encompassing view of the model's performance for false positives, false negatives, true positives, and true negatives.

**Using Humans to check:**

Sometimes the ML Model can be prone to make mistakes as its not 100% accurate, so using humans / doctors to validate and test the data would be a huge ideal.

## Compare Expected / Correct results ;

**Confusion matrix**

Expected Outcome: A confusion matrix ought to display a low proportion of false positives and false negatives and a large number of genuine positives and true negatives.
Correct Result: Examine the discrepancy between the actual and anticipated confusion matrices. Find any trends in the incorrect categories.

**Precision and Recall:**

Expected Outcome: It would be ideal to have high recall and precision values, particularly when identifying fractures.
Correct Result : Verify that the recall and precision figures match the predictions. Recall is the capacity of the model to catch all positive instances, whereas precision is concerned with the accuracy of positive predictions.
